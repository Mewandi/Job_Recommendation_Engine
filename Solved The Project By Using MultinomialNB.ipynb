### Let's combine all datas to make one

```python
job_col_df.rename(columns={'skills': 'required_skills'}, inplace=True)
emp_cols_df.rename(columns={'skills': 'employee_skills'}, inplace=True)
```

```python
total_df = pd.concat([job_col_df, emp_cols_df])
```

```python
total_df = pd.DataFrame(total_df)
```

```python
100 * total_df.isnull().sum() / len(total_df)
```

```python
for i in total_df:
    total_df[i] = total_df[i].fillna(" ")
```

```python
100 * total_df.isnull().sum() / len(total_df)
```

```python
total_df.head(2)
```

```python
total_df['combined_cols'] = (   
                                total_df['description'] + " " +
                                total_df['status'] + " " +
                                total_df['required_skills'] + " " +
                                total_df['functionalArea'] + " " +
                                total_df['employee_skills']
                             )
```

```python
total_df['combined_cols'] = total_df['combined_cols'].fillna(" ")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

x_train = cv.fit_transform(total_df['combined_cols'])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(total_df['jobTitle'])
y_train
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
# y_pred = model.predict(x_test)
```

```python
data = {
    'functionalArea': ['Computer Science'],
    'description': ['I understand computer languages and also, I solved many projects in AI and Machine Learning'],
    'status': ['Active'],
    'required_skills': ['computer languages'],
    'employee_skills': ['I am a programmer and I can do whatever you want'],
}
data = pd.DataFrame(data)

# combine columns in the same way as training
data['combined_cols'] = (
    data['description'] + " " +
    data['status'] + " " +
    data['required_skills'] + " " +
    data['functionalArea'] + " " +
    data['employee_skills']
)
data = data.fillna(" ")

# ✅ only transform, do NOT fit again
X_test = cv.transform(data['combined_cols'])

# ✅ predict
y_pred = model.predict(X_test)
print("Encoded prediction:", y_pred)

# Decode back to string
predicted_jobTitle = label_encoder.inverse_transform(y_pred)
print("Predicted jobTitle:", predicted_jobTitle[0])
```
